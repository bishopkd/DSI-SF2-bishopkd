```python

ath_file = '../../../../DSI-SF2-bishopkd/datasets/highest_paid_athletes/Athletes.csv'
dfa = pd.read_csv(ath_file)

dfa.head(1)
```

```python
dfa.rename(columns={
        'Total Pay':'total_pay',
        'Salary/Winnings':'salary',
        'Year of birth':'birth_year',
        'Height (cm)':'heigh',
        'Birth Date':'birth_date',
        'Place of Birth':'birth_place'
    }, inplace=True)
dfa.head(1)
```

```python
dfa_sub = dfa[['total_pay','salary','Endorsements','birth_date','heigh']]
dfa_sub.describe()
```

```python

def clean(col):
    col = col.replace('$','').replace(',','')
    return float(col)
    

dfa['total_pay'] = dfa['total_pay'].map(clean)
dfa['salary'] = dfa['salary'].map(clean)
dfa['Endorsements'] = dfa['Endorsements'].map(clean)

dfa.head(2)
```

```python
#top_mask = (dfa.total_pay

dfa_top_short = dfa.groupby('Name').head(10).sort(['heigh'], ascending=[True])
dfa_top_short.head(10)
```

```python
def feet_convert(cm):
    return cm * 0.03281

dfa['heigh'] = dfa['heigh'].map(feet_convert)
```
