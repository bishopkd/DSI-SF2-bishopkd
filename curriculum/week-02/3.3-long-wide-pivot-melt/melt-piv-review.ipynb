```python
import pandas as pd
import numpy as np
```

```python
beers = {
    'name':['coors','bud','natural light','keystone ice',
            'sierra nevada', 'sam adams', 'rolling rock',
            'odouls',
            'pbr','stella','chimay','magnolia','21amendment'],
    'class':['crap','crap','crap','crap',
             'mid','mid','mid',
             'notabeer',
             'pretentious','pretentious','pretentious','pretentious','pretentious']
}
beers['price'] = np.concatenate([np.array([1.5, 1.7, 1.2, 1.2]),
                                np.array([2., 1.9, 2.1]),
                                np.array([3.]),
                                np.array([0.5,3.5, 10., 15.,2.])])
beers['rating'] = np.random.normal(5, 2, size=13)

for k, v in beers.items():
    print k, len(v)
beers = pd.DataFrame(beers)
beers
```

```python
beers_long = pd.melt(beers, id_vars=['name'])
```

```python
beers_long.sort_values('name',axis=0)
```

```python
# index is group by
# can't have std dev of one item, etc
beer_class_summary = pd.pivot_table(beers, index='class', values=['rating','price'], aggfunc=[np.mean, np.std])
beer_class_summary
```

```python
def first_item(series):
    item = series.iloc[0]
    return item

new_beers_wide = pd.pivot_table(beers_long, index='name', values='value', columns=['variable'], aggfunc=first_item)
new_beers_wide.loc['21amendment', ['price','class']]
new_beers_wide.reset_index()
#new_beers_wide.reset_index(drop=True) -- you will lose original index
```
